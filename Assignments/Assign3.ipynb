```python
Przykład nr 24

clf;
x = (1:5)';
subplot(2,2,1);
plot(x, rand(numel(x)), 'r'); # Zmiana koloru linii na czerwony
legend(cellstr(num2str(x)), "location", "northwestoutside");
subplot(2,2,2);
plot(x, rand(numel(x)), 'g'); # Zmiana koloru linii na zielony
legend(cellstr(num2str(x)), "location", "northeastoutside");
subplot(2,2,3);
plot(x, rand(numel(x)), 'b'); # Zmiana koloru linii na niebieski
legend(cellstr(num2str(x)), "location", "southwestoutside");
subplot(2,2,4);
plot(x, rand(numel(x)), 'm'); # Zmiana koloru linii na magenta
legend(cellstr(num2str(x)), "location", "southeastoutside");

set(gcf, 'Color', [0.95, 0.95, 0.95]);
sgtitle("Wykresy w podwykresach z legendami na zewnątrz");

# Zmieniono kolor linii dla wszystkich wykresów na różowy ('r'), zielony ('g'), niebieski ('b') i magenta ('m').
# Dodano tytuł globalny (sgtitle) nad wszystkimi podwykresami.
# Zmieniono kolor tła wykresu na jasnoszary.
# Wszystkie wykresy posiadają legendy umieszczone na zewnątrz obszaru wykresu.
```
